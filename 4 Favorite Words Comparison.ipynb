```python
import pandas as p
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from textblob import TextBlob
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

```python
input_directory1 = "C:/Users/BLAHBLAHBLAH/Desktop/tweets_from_seanhannity_followers_cleaned.csv"
input_directory2 = "C:/Users/BLAHBLAHBLAH/Desktop/tweets_from_maddow_followers_cleaned.csv"

person1 = "Sean Hannity"
person2 = "Rachel Maddow"
```

```python
data_a = p.read_csv(input_directory1, dtype = {'tweets': str})
data_a.head(5)
```

```python
data_a0 = data_a
data_a0.tweets = data_a.tweets.apply(str)
data_a0.head(5)
```

```python
data_b = p.read_csv(input_directory2, dtype = {'tweets': str})
data_b.head(5)
```

```python
data_b0 = data_b
data_b0.tweets = data_b.tweets.apply(str)
data_b0.head(5)
```

```python
data_both = p.DataFrame({
    'name': [person1, person2],
    'tweets_all': [" ".join(data_a0.tweets), " ".join(data_b0.tweets)]
})
data_both.head(5)
```

```python
cv = CountVectorizer(stop_words = 'english')
data_cv = cv.fit_transform(data_both.tweets_all)
data_dtm = p.DataFrame(
    data_cv.toarray(),
    columns = cv.get_feature_names()
)
data_dtm.index = data_both.name
#data2_dtm.to_pickle('dtm.pkl')
data_dtm = data_dtm.transpose()
data_dtm.head(5)
```

```python
data_vocabulary_comparison = p.DataFrame({
    'word_ratio': data_dtm.iloc[:,0]/data_dtm.iloc[:,1],
    'word_difference': data_dtm.iloc[:,0] - data_dtm.iloc[:,1],
    'count_Republicans': data_dtm.iloc[:,0],
    'count_Democrats': data_dtm.iloc[:,1]
})
```

```python
data_vocabulary_comparison.sort_values(by = ['word_ratio', 'word_difference']).query(
    'count_Democrats > 24 & count_Republicans > 0'
).iloc[0:60,]
```

```python
data_vocabulary_comparison.sort_values(by = ['word_ratio', 'word_difference'], ascending = False).query(
    'count_Republicans > 25 & count_Democrats > 0'
).iloc[0:60,]
```
